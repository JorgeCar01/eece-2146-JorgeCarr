```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 20)
f = 3 * (x ** 3) + 4 * (x ** 2) - 2

N = len(x)
Dx = np.zeros((N, N))
d = np.ones(N)
```

```python
np.fill_diagonal(Dx[1:], -1)
np.fill_diagonal(Dx[:, 1:], d)

dx = x[1] - x[0]
Dx = Dx / (2 * dx)

fp_numerical = np.matmul(Dx, f)
fp_numerical[0] = fp_numerical[-1] = None  # set to None since it leads to an unexpected spike since central finite-difference does not provide a definition for the boundaries
fp_analytical = 9 * (x ** 2) + 8 * x
```

```python
plt.figure(dpi=100)
plt.plot(x, f, '.-', label='f(x)')
plt.plot(x, fp_analytical, '--', label="analytical derivative")
plt.plot(x, fp_numerical, 'g.-', markersize=4, linewidth=0.5, label='numerical derivative')
plt.legend()
plt.grid()
plt.xlabel("x")

#ax = plt.gca()
#ax.set_ylim([-1.5, 1.5])

plt.show()
```
