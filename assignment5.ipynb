```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the function
def f(x):
    return x**4 - x - 10

# Generate x values
x = np.linspace(-2, 2, 400)

# Compute y values
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = $x^4 - x - 10$')
plt.title('Plot of f(x) = $x^4 - x - 10$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='red',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.ylim(-12, 10)
plt.show()
```

```python
def bisection(a, b, epsilon):
    if f(a) * f(b) > 0:
        return None
    prev_midpoint = a
    while True:
        midpoint = (a + b) / 2.0
        delta_x = abs(midpoint - prev_midpoint) / abs(prev_midpoint)
        print(f"a = {a}, b = {b}, mid = {midpoint}, f(a) = {f(a)}, f(m) = {f(midpoint)}, f(a)f(m) = {f(a) * f(midpoint)}, delta_x = {delta_x}")
        if f(midpoint) == 0 or delta_x < epsilon:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        
        prev_midpoint = midpoint
    return (a + b) / 2.0
```

```python
a = 1.5
b = 2.0

epsilon = 0.02

root = bisection(a, b, epsilon)
if root is not None:
    print(f"The root found in the interval [{a}, {b}] is approximately {root:.5f}.")
else:
    print("The bisection method is not applicable in this interval.")
```
