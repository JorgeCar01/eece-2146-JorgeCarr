```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
```

```python
# Left Riemann Sum
print('Left Riemann Sum')

LeftAreas = []
N = [5, 10, 100]
a = 0
b = np.pi/2
for n in N:
    x = np.linspace(a, b, n)
    y = np.cos(x)
    fig, ax = plt.subplots()
    plt.plot(x, y, 'r.-')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axvline(0, c='k')
    plt.axhline(0, c='k')
    plt.axvline(0, c='b')
    plt.axhline(0, c='b')
    A = 0
    dx = x[1] - x[0]
    for i in range(n):
        xi = x[i]
        yi = y[i]
        ax.add_patch(Rectangle((xi, 0), dx, yi, linewidth=0.5, edgecolor='k', facecolor='gray', alpha=0.3))

        A += dx * yi
    LeftAreas.append(A)
    print(f"Number of points: {n}")
    print(f"Total area = {A}")

    plt.show()
```

```python
# Left Riemann Sum
print('Left Riemann Sum')

RightAreas = []
N = [5, 10, 100]
a = 0
b = np.pi/2
for n in N:
    x = np.linspace(a, b, n + 1)
    y = np.cos(x)
    fig, ax = plt.subplots()
    plt.plot(x, y, 'r.-')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axvline(0, c='k')
    plt.axhline(0, c='k')
    plt.axvline(0, c='b')
    plt.axhline(0, c='b')
    A = 0
    dx = (b - a) / n
    for i in range(1, n + 1):
        xi = x[i]
        yi = y[i]
        ax.add_patch(Rectangle((xi - dx, 0), dx, yi, linewidth=0.5, edgecolor='k', facecolor='gray', alpha=0.3))

        A += dx * yi
    RightAreas.append(A)
    print(f"Number of points: {n}")
    print(f"Total area = {A}")

    plt.show()
```

```python
import sympy as sp

x = sp.symbols('x')

f = sp.cos(x)

integral = sp.integrate(f, (x, 0, sp.pi/2))
print(f"Actual integral: {integral}\n")
for leftA, rightA, n in zip(LeftAreas, RightAreas, N):
    deltaLeft = abs(leftA - integral)
    deltaRight = abs(rightA - integral)
    print(f"Number of steps: {n}")
    print("Using Left Riemann Sum:")
    print(f"\tIntegral from Riemann: {leftA}\n\tDifference from actual: {deltaLeft}")
    print("Using Right Riemann Sum:")
    print(f"\tIntegral from Riemann: {rightA}\n\tDifference from actual: {deltaRight}")
    if deltaRight < deltaLeft:
        comp = "Right was better\n"
    elif deltaRight > deltaLeft:
        comp = "Left was better\n"
    else:
        comp = "Both were the same\n"
    print(comp)

# Based on the results Right Riemann was better
```
