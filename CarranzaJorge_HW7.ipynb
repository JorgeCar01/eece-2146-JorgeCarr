```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(2 * x)
```

```python
def df(x):
    return 2 * np.cos(2 * x)
```

```python
x = np.linspace(-5, 5, 400)
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = sin(2x)')
plt.title('Plot of f(x) = sin(2x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='red',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.axvline(x=2, color='purple', linewidth =1.5, label='x = 4')
plt.legend(loc='upper right')
plt.ylim(-1.1, 1.1)
plt.show()
```

```python
def newton_raphson(xi, tol, max_iter=100):
    prev_x = xi
    for i in range(max_iter):
        print(f"i = {i}")       
        if df(prev_x) == 0:
            print("Derivative is zero. No solution found.")
            return None
        x = prev_x - f(prev_x)/df(prev_x)
        delta_x = abs(x - prev_x)
        print(f"x_i = {prev_x}, f(x_i) = {f(prev_x)}, f'(x_i) = {df(prev_x)}, \nx_i+1 = {x}, delta_x = {delta_x}\n")
        if delta_x < tol:
            return x        
        prev_x = x
        
    print(f"Newton-Raphson did not converge after {max_iter} iterations.")
    return None
```

```python
xi = 2

# Tolerance level
tolerance = 0.01

root = newton_raphson(xi, tolerance)
if root is not None:
    print(f"The root found using the Newton-Raphson method is approximately {root:.5f}.\nf(root) = {f(1.5708)}")
```
